## 1. Introduction & Problem Formulation <a id='1-introduction'></a>

### 1.1 Motivation

Imagine you are standing at the **Rila Monastery** and you want to reach the **Seven Rila Lakes**.
There are multiple possible paths. Which one should you take?

A naive answer might be: *"the shortest one"*. But in mountain terrain, this is rarely the right answer.
A short path with steep elevation gain can take **twice as long** — and be far more exhausting —
than a longer path with gentler slopes.

This is the core problem this project addresses:

> **How do we define and find the "optimal" route in the mountains — and does the answer change depending on what we optimize for?**

---

### 1.2 Real-World Significance

This problem has practical applications in several domains:

- 🏃 **Trail running & hiking** — athletes and hikers want to plan routes based on time and effort, not just distance
- 🚁 **Mountain rescue operations** — rescue teams need the fastest, safest route to a specific location
- 🗺️ **Navigation apps** — apps like Komoot and Wikiloc need to suggest routes adapted to user fitness
- 🌍 **Environmental studies** — researchers tracking wildlife or vegetation need elevation-aware pathfinding

### 1.5 Key Question

> *Is the shortest path always the best path in the mountains?*

We will answer this by comparing three definitions of "optimal":

- 📏 **Shortest distance** — minimize total kilometers
- ⛰️ **Least elevation gain** — minimize total ascent in meters  
- ⏱️ **Fastest time** — minimize estimated hiking time using **Naismith's Rule**
